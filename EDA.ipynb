```python
# Function to read text file

def readTextFile(filePath):
    with open(filePath, 'r', encoding='utf-8') as file:
        text = file.read()
    return text
```

```python
# try to load the file

text = readTextFile('RUU Cipta Kerja.txt')
text
```

```python
# clean a bit

import re

def parseText(text):
    pattern =  re.compile(r'(Pasal \d+)(.*?)(Cukup jelas\.)', re.DOTALL)
    matches =  pattern.findall(text)

    data = []
    for match in matches:
        pasal = match[0]
        content = match[1].strip()
        data.append(pasal)
        data.append(content)

    return data
```

```python
text = parseText(text = text)
```
